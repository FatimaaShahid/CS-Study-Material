## Implementing Gradient Descent Algorithm

```python
def gradient_descent(derivative_func, learning_rate=0.02, 
                     initial_value=0.5, precision=0.0001, max_iter=300):
    t=initial_value
    t_list=[]

    for i in range(1,max_iter+1):
        t_list.append(t)
        t = t - learning_rate * derivative_func(t)
        
        if abs(learning_rate * derivative_func(t))<=precision:
        # if derivative_func(t)==0:
            break
        print(i,"d(f(t))",derivative_func(t),"l",learning_rate * derivative_func(t),"t",t)
        
    return t,t_list,i
```

### Partial derivatives for $$ f(t1,t2)=\frac{1}{3^{-t1^2-t2^2}+1} $$ w.r.t. t1 and t2

$$ \frac{\partial f(t1,t2)}{\partial dt1} = \frac{2t1.ln(3).3^{-t1^2-t2^2}}{({3^{-t1^2-t2^2}}+1)^2} $$
$$ \frac{\partial f(t1,t2)}{\partial dt2} = \frac{2t2.ln(3).3^{-t1^2-t2^2}}{({3^{-t1^2-t2^2}}+1)^2} $$

Use online tools for finsing derivatives.

```python
#Implementing partial derivatives
def pdfx(t1,t2):
    return 2*t1*log(3)*(3**(-t1**2-t2**2))/(3**(-t1**2-t2**2)+1)**2
def pdfy(t1,t2):
    return 2*t2*log(3)*(3**(-t1**2-t2**2))/(3**(-t1**2-t2**2)+1)**2
```

```python
#Custom implementation of gradient descent for two parameters
learning_rate = 0.1
max_iter = 20000
initial_t1 = 1.8
initial_t2=1.5
precision=0.000001

t1_old=initial_t1
t2_old=initial_t2

t1_list=[]
t2_list=[]
for i in range (1,max_iter+1):
    t1_new = t1_old - learning_rate * pdfx(t1_old,t2_old)
    t2_new = t2_old - learning_rate * pdfy(t1_old,t2_old)
    
    t1_list.append(t1_old)
    t2_list.append(t2_old)

    t1_old=t1_new
    t2_old=t2_new
```

```python
#a
import numpy as np

# Function
def f(t1, t2, t3):
    return t1**2 + t2**2 + t3**2

# Partial derivatives
def pdf_t1(t1, t2, t3):
    return 2 * t1

def pdf_t2(t1, t2, t3):
    return 2 * t2

def pdf_t3(t1, t2, t3):
    return 2 * t3
```

```python
def gradient_descent_3d(pdf_t1, pdf_t2, pdf_t3, f,
                        learning_rate=0.02,
                        max_iter=5000,
                        precision=1e-6,
                        t1_init=1.5,
                        t2_init=2.0,
                        t3_init=-1.0):

    t1, t2, t3 = t1_init, t2_init, t3_init

    for i in range(1, max_iter+1):
        # Compute gradients
        grad_t1 = pdf_t1(t1, t2, t3)
        grad_t2 = pdf_t2(t1, t2, t3)
        grad_t3 = pdf_t3(t1, t2, t3)

        # Update parameters
        t1_new = t1 - learning_rate * grad_t1
        t2_new = t2 - learning_rate * grad_t2
        t3_new = t3 - learning_rate * grad_t3

        # Check convergence
        grad_norm = np.sqrt(grad_t1**2 + grad_t2**2 + grad_t3**2)
        if grad_norm < precision:
            print(f"Converged in {i} iterations")
            return t1_new, t2_new, t3_new, f(t1_new, t2_new, t3_new)

        # Update for next iteration
        t1, t2, t3 = t1_new, t2_new, t3_new

    print("Reached max iterations without meeting precision")
    return t1, t2, t3, f(t1, t2, t3)

# Run gradient descent
t1_final, t2_final, t3_final, cost_final = gradient_descent_3d(pdf_t1, pdf_t2, pdf_t3, f)

print("\n--- Final Results ---")
print(f"t1 = {t1_final}, t2 = {t2_final}, t3 = {t3_final}")
print(f"Cost = {cost_final}")
```
